```python
import matplotlib.pyplot as plt

data = [
    ["300-400", 14],
    ["400-500", 56],
    ["500-600", 60],
    ["600-700", 86],
    ["700-800", 74],
    ["800-900", 62],
    ["900-1000", 48]
]

# Separate data into bin labels and counts
bin_labels = []
counts = []

for item in data:
    bin_labels.append(item[0])
    counts.append(item[1])


# Create a histogram
plt.figure(figsize=(8, 6))
plt.bar(bin_labels, counts, width=0.5, color='skyblue')
plt.xlabel('Bin Range')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
