```python
import pandas as pd
import numpy as np

df=pd.read_csv('C:\\Users\\avinash\\Downloads\\q11.csv')
print(df)

from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer=imputer.fit(df[['Salary']])
df[['Salary']]=imputer.transform(df[['Salary']])
df

imput=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imput=imput.fit(df[['Gender','Country','Company']])
df[['Gender','Country','Company']]=imput.transform(df[['Gender','Country','Company']])
df
```
