```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = [[3, 12], [0, 0], [12, 9], [0, 4], [2, 15], [0, 2], [26, 6], [0, 10], [7, 27], [5, 15], [5, 5], [2, 9], [1, 1], [1, 14], [2, 2]]
field_1 = []
field_2 = []

for pair in data:
    field_1.append(pair[0])
    field_2.append(pair[1])


# Calculate IQR and identify outliers for Field 1
q1_field_1 = np.percentile(field_1, 25)
q3_field_1 = np.percentile(field_1, 75)
iqr_field_1 = q3_field_1 - q1_field_1
lower_bound_field_1 = q1_field_1 - 1.5 * iqr_field_1
upper_bound_field_1 = q3_field_1 + 1.5 * iqr_field_1

outliers_field1 = []
outliers_field2 = []

for x in field_1:
    if x < lower_bound_field_1 or x > upper_bound_field_1:
        outliers_field1.append(x)

for y in field_2:
    if y < lower_bound_field_2 or y > upper_bound_field_2:
        outliers_field2.append(y)


# Calculate IQR and identify outliers for Field 2
q1_field_2 = np.percentile(field_2, 25)
q3_field_2 = np.percentile(field_2, 75)
iqr_field_2 = q3_field_2 - q1_field_2
lower_bound_field_2 = q1_field_2 - 1.5 * iqr_field_2
upper_bound_field_2 = q3_field_2 + 1.5 * iqr_field_2



# Create box plots for Field 1 and Field 2
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.boxplot(field_1)
plt.title('Box Plot for Field 1')
plt.subplot(1, 2, 2)
plt.boxplot(field_2)
plt.title('Box Plot for Field 2')
plt.show()

# Create a scatterplot for the data
plt.figure(figsize=(6, 6))
plt.scatter(field_1, field_2)
plt.title('Scatter Plot')
plt.xlabel('Field 1 (X values)')
plt.ylabel('Field 2 (Y values)')
plt.grid(True)
plt.show()

print("Outliers in Field 1:", outliers_field1)
print("Outliers in Field 2:", outliers_field2)
```
